For the the time being, we have used the Brillouin function for a $\textbf{paramagnet}$. Meaning:

\begin{equation}
B_J(x) = \frac{2J + 1}{2J} \coth \left ( \frac{2J + 1}{2J} x \right )
                - \frac{1}{2J} \coth \left ( \frac{1}{2J} x \right )
\end{equation}

Magnetization, normalized by the saturation magnetization, is given by:

\begin{equation}
\frac{M}{M_{\rm S}} =  B_J(x)
\end{equation}
In the calculations I assume Manganese atoms in Mn$^{2+}$ state having total angular momentum $J=5/2$.

# Anisotropy studies

Results in this section were obtained for two ${\rm Pb_{0.16}Sn_{0.74}Mn_{0.10}Te}$ samples with the same mass $m=23.9\,mg$ and the same geometry. The samples were measured in the same magnetometer, which was previously de-gaussified. We assume that remanent field in the magnet was the same, so we might assume high experimental accuracy.

A peculiar behavior of magnetization as a function of magnetic field appears from the plots above. Seemingly, the magnetization for an easy direction is higher in up to 1 kOe. Interestingly, Magnetization is reaching higher value for the higher field regime. Consistently, at every measured tempeartue, magnetization for the hard direction exceeds the one measured on an easy axis. The same behavior was measured by Eggenkamp in his PhD thesis. Ironically the author took it as an experimental inaccuracy. The picture from Eggakamp's thesis is presented below.
